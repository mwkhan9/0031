```python
# Importing necessary libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

#Insert path here
path = "Data/"  
survey = path + 'survey.csv'
answers = path + 'answers.csv'
questions = path + 'questions.csv'

#pandas df's:
#survey_df = pd.read_csv(survey)

questions_df = pd.read_csv(questions)
questions_df = questions_df.rename(columns={'questionid': 'QuestionID'})

answers_df = pd.read_csv(answers)
answers_df = answers_df.rename(columns={'SurveyID': 'Year'})
```

## Merge into a single table

```python
df = pd.merge(answers_df, questions_df, on='QuestionID')
df
df.to_csv(path+"", sep='\t')
```

## See variations in responses dependent on country of employment

```python
countries = df[df['QuestionID'] == 3]['AnswerText'].unique()
#countries
```

```python
#filter responses for question 3 only
df_question_3 = df[df['QuestionID'] == 3]

## Will split into groups in uk, usa and elsewhere

# Define the groups (USA, UK, and everywhere else)
usa_group = ['United States of America','United States']
uk_group = ['United Kingdom']

countries = df[df['QuestionID'] == 3]['AnswerText'].unique()

everywhere_else_group = [x for x in countries if x not in usa_group]
everywhere_else_group = [x for x in everywhere_else_group if x not in uk_group]


# Create a new column for the groups based on AnswerText
df_question_3['Group'] = df_question_3['AnswerText'].apply(lambda x: 'USA' if x in usa_group else ('UK' if x in uk_group else 'Everywhere Else'))

# Group by Year and Group, then count unique users
result = df_question_3.groupby(['Year', 'Group'])['UserID'].nunique().reset_index()

# Pivot result for better readability 
result_pivot = result.pivot(index='Year', columns='Group', values='UserID').fillna(0)

# Print or display the result
print(result_pivot)
```

```python
import pandas as pd

# Assuming your DataFrame is named df
# Filter the DataFrame for QuestionID 3
df_question_3 = df[df['QuestionID'] == 3]

# Define the groups (USA, UK, and everywhere else)
usa_group = ['United States of America','United States']
uk_group = ['United Kingdom']

countries = df[df['QuestionID'] == 3]['AnswerText'].unique()

everywhere_else_group = [x for x in countries if x not in usa_group]
everywhere_else_group = [x for x in everywhere_else_group if x not in uk_group]


# Create a new column for the groups based on AnswerText
df_question_3['Group'] = df_question_3['AnswerText'].apply(lambda x: 'USA' if x in usa_group else ('UK' if x in uk_group else 'Elsewhere'))

# Filter the DataFrame for QuestionID 10
df_question_10 = df[df['QuestionID'] == 10]

# Merge the two DataFrames on UserID
merged_df = pd.merge(df_question_3, df_question_10[['UserID', 'AnswerText']], on='UserID', how='inner', suffixes=('_q3', '_q10'))

# Group by Year and Group, then count unique users and responses for QuestionID 10
result = merged_df.groupby(['Year', 'Group', 'AnswerText_q10'])['UserID'].nunique().unstack(fill_value=0).reset_index()

# Print or display the result
print(result)
```

```python
result["I don't know"] = result["I don't know"].astype(int) + result["Don't know"].astype(int)
result = result.drop(columns=["Don't know"])
#result
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Assuming your DataFrame is named result
# You can modify the code accordingly based on the actual result DataFrame

# Plot bar graphs for each group
for group in ['USA', 'UK', 'Elsewhere']:
    group_data = result[result['Group'] == group]
    
    # Plot bar graph for each country_group and each responses
    plt.figure(figsize=(10, 6))
    plt.bar(group_data['Year'], group)
    
    # Plot bar graph for QuestionID 10 counts
    for answer_text in group_data.columns[3:]:
        #plt.bar(group_data['Year'], group_data[answer_text], label=answer_text)
        plt.bar(group_data['Year'], group_data[answer_text], label=answer_text)

    plt.title(f'Counts for Group: {group}')
    plt.xlabel('Year')
    plt.ylabel('Count')
    plt.legend()
    plt.show()
```

```python
# Print values for each group
for group in ['USA', 'UK', 'Elsewhere']:
    group_data = result[result['Group'] == group]
    
    print(f'\nCounts for Group: {group}\n')
    print(group_data.to_string(index=False))
```
